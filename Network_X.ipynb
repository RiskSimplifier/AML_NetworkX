```python
import networkx as nx
import nxviz as nv
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edge('412359876032451','1234567890123450', weight=9)
G.add_edge('5096328741509830','5432109876543210', weight=2)
G.add_edge('6789054321890760','9876543210123450', weight=1)
G.add_edge('2143568970325870','8901234567890120', weight=8)
G.add_edge('8940763212356780','4567890123456780', weight=7)
G.add_edge('4567890123456780','1234567890123450',weight=5)
G.add_edge('8901234567123450','1234567890123450',weight=1)
G.add_edge('3456789012345670','4321098765432100', weight=2)
G.add_edge('1234567890123450','8901234567890120',weight=6)
G.add_edge('5678901234567890','1234567890123450',weight=4)
```

```python
#Create a circular layout
pos = nx.circular_layout(G)

# Increase figure size
plt.figure(figsize=(10, 8))



# Extract edge weights
edge_weights = [edge[2]['weight'] for edge in G.edges(data=True)]

# Normalize edge weights for display size
edge_weights_normalized = [float(i) / sum(edge_weights) * 10 for i in edge_weights]

# Draw the graph with circular layout, labels, and edge weights
nx.draw(G, pos, with_labels=True, font_weight='bold', font_size=9, node_size=300, node_color='skyblue', font_color='black', width=edge_weights_normalized, edge_color='gray')

# Add labels to edges with weights
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G,pos, edge_labels=edge_labels)

# Show the plot
plt.show()
```

```python
import pandas as pd
# Calculate degree centrality
degree_centrality = nx.degree_centrality(G)

# Create a DataFrame from the degree centrality dictionary
df = pd.DataFrame(list(degree_centrality.items()), columns=['Node', 'Degree Centrality'])
df['Node'] = df['Node'].astype(str)  # Convert node labels to string for better display

# Sort the DataFrame by 'Node' in ascending order
df = df.sort_values(by='Node', ascending=True)

# Display the DataFrame
print(df)
```

```python
import networkx as nx
import pandas as pd

def calculate_betweenness_centrality(graph):
    # Calculate betweenness centrality
    betweenness_centrality = nx.betweenness_centrality(graph, weight='weight')

    # Create a DataFrame to display the results
    df = pd.DataFrame(list(betweenness_centrality.items()), columns=['Node', 'Betweenness Centrality'])

    # Sort the DataFrame by betweenness centrality in descending order
    df = df.sort_values(by='Betweenness Centrality', ascending=False).reset_index(drop=True)

    return df



# Calculate and display betweenness centrality
result_df = calculate_betweenness_centrality(G)

# Display the modified DataFrame
print(result_df)
```

```python
import community.community_louvain as community_louvain  # Louvain algorithm is part of the community package

# Create a Pandas DataFrame for node attributes
df_node_attributes = pd.DataFrame(data)
df_node_attributes.set_index('Node', inplace=True)

# Add node attributes to the graph
for node, attributes in df_node_attributes.iterrows():
    G.nodes[node]['EmploymentStatus'] = attributes['EmploymentStatus']
    G.nodes[node]['RiskCategory'] = attributes['RiskCategory']

# Apply Louvain algorithm for community detection
partition = community_louvain.best_partition(G)

# Add the community information to the graph as a node attribute
nx.set_node_attributes(G, partition, 'community')

# Display the communities
communities = {}
for node, comm in partition.items():
    if comm not in communities:
        communities[comm] = [node]
    else:
        communities[comm].append(node)

print("Groups of Customers:")
for comm, nodes in communities.items():
    print(f"Groups of Customers {comm}: {nodes}")
```

```python
# Add the community information to the graph as a node attribute
nx.set_node_attributes(G, partition, 'community')

# Create a non-interactive plot with labels and features
pos = nx.spring_layout(G, k=0.9)  # Adjust the 'k' value to control node distance

# Extract node colors based on community
node_colors = [partition[node] for node in G.nodes()]


# Extract labels and features for nodes
labels = {node: node for node in G.nodes()}
employment_status = nx.get_node_attributes(G, 'EmploymentStatus')
risk_category = nx.get_node_attributes(G, 'RiskCategory')
weights = nx.get_edge_attributes(G, 'weight')

# Draw the graph
plt.figure(figsize=(12, 10))
nx.draw(G, pos=pos, node_color=node_colors, cmap=plt.cm.Set1, with_labels=True, labels=labels, font_size=8)

# Display node features as text with increased spacing
for node, (x, y) in pos.items():
    text = f"\nEmp: {employment_status[node]}\nRisk: {risk_category[node]}"
    plt.text(x, y - 0.1, text, fontsize=8, ha='center')  # Adjust the y-coordinate for increased spacing

# Display edge weights as labels
edge_labels = {(edge[0], edge[1]): weights.get(edge, '') for edge in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Display the plot
plt.show()
```
